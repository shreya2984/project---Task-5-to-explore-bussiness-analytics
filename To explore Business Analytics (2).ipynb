```python
# Profit Percentage of top10 states
```

```python
profit_percentage_of_top_10_states= df.groupby('State').Profit_Percentage.sum().nlargest(n=10)
print(profit_percentage_of_top_10_states)
```

```python
plt.figure(figsize=(12,9))
profit_percentage_of_top_10_states.plot(kind='bar', colormap='flare')
plt.title('Top 10 states by Profit Percentage',fontsize=20)
plt.xlabel('States',fontsize=15)
plt.ylabel('Total Profit Percentage',fontsize=15)
plt.tight_layout()
```

```python
# Profit Percentage of below 10 states
```

```python
profit_percentage_of_below_10_states= df.groupby('State').Profit_Percentage.sum().nsmallest(n=10)
print(profit_percentage_of_below_10_states)
```

```python
plt.figure(figsize=(12,9))
profit_percentage_of_below_10_states.plot(kind='bar', colormap='flare')
plt.title('Below 10 states by Profit Percentage',fontsize=20)
plt.xlabel('States',fontsize=15)
plt.ylabel('Total Profit Percentage',fontsize=15)
plt.tight_layout()
```

```python
import plotly as py

from plotly.offline import download_plotlyjs, init_notebook_mode, plot, iplot

import plotly.graph_objs as go

init_notebook_mode(connected=True)
```

```python
data_profit = dict(
    type = 'choropleth',
    locations = df['state_code'],
    locationmode = 'USA-states',
    colorscale = 'magenta',
    z = df['Sales'].astype(float),
    colorbar = {'title' : 'Sales by State'}

)

layout = dict(geo={'scope':'usa'})

choromap = go.Figure([data_profit], layout)

iplot(choromap)
```
